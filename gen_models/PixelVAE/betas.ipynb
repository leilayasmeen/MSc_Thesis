```python
# Author: Jake VanderPlas
# License: BSD
#   The figure produced by this code is published in the textbook
#   "Statistics, Data Mining, and Machine Learning in Astronomy" (2013)
#   For more information, see http://astroML.github.com
#   To report a bug or issue, use the following forum:
#    https://groups.google.com/forum/#!forum/astroml-general
import numpy as np
from scipy.stats import beta
from matplotlib import pyplot as plt
```

```python

#----------------------------------------------------------------------
# This function adjusts matplotlib settings for a uniform feel in the textbook.
# Note that with usetex=True, fonts are rendered with LaTeX.  This may
# result in an error if LaTeX is not installed on your system.  In that case,
# you can set usetex to False.
from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=8, usetex=True)
```

```python

#------------------------------------------------------------
# Define the distribution parameters to be plotted
alpha_values = [1.0, 0.6, 0.2, 0.1]
beta_values = [1.0, 0.6, 0.2, 0.1]

# style
plt.style.use('seaborn-whitegrid')
 
# create a color palette
palette = plt.get_cmap('Set1')


x = np.linspace(0, 1, 1002)[1:-1]

#------------------------------------------------------------
# plot the distributions
fig, ax = plt.subplots(figsize=(5, 3.75))
i = 0

for a, b in zip(alpha_values, beta_values):
    dist = beta(a, b)

    plt.plot(x, dist.pdf(x), c=palette(i),
             label=r'Beta ($%.1f,%.1f$)' % (a, b))
    i = i + 1

plt.xlim(0, 1)
plt.ylim(0, 3)

plt.xlabel('$\lambda$',fontsize=12)
plt.ylabel(r'$p(\lambda|\alpha,\alpha)$',fontsize=12)
plt.title(r'Beta $(\alpha,\alpha)$ distribution by value of $\alpha$',fontsize=12)

plt.legend(loc='best',fontsize=12)
#plt.show()

# Plot
plt.savefig('/Users/wildflowerlyi/Desktop/Plots/betas.png')

```
