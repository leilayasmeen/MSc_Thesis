```python
# Author: Jake VanderPlas
# License: BSD
#   The figure produced by this code is published in the textbook
#   "Statistics, Data Mining, and Machine Learning in Astronomy" (2013)
#   For more information, see http://astroML.github.com
#   To report a bug or issue, use the following forum:
#    https://groups.google.com/forum/#!forum/astroml-general
import numpy as np
from scipy.stats import beta
from matplotlib import pyplot as plt
```

```python

#----------------------------------------------------------------------
# This function adjusts matplotlib settings for a uniform feel in the textbook.
# Note that with usetex=True, fonts are rendered with LaTeX.  This may
# result in an error if LaTeX is not installed on your system.  In that case,
# you can set usetex to False.
from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=8, usetex=False)
```

```python

#------------------------------------------------------------
# Define the distribution parameters to be plotted
alpha_values = [1.0, 0.6, 0.2, 0.1]
beta_values = [1.0, 0.6, 0.2, 0.1]

# style
plt.style.use('seaborn')
 
# create a color palette
palette = plt.get_cmap('Set1')

plt.rc('font', size=15)          # controls default text sizes
plt.rc('axes', titlesize=15)     # fontsize of the axes title
plt.rc('axes', labelsize=15)
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plt.rc('legend', fontsize=15)    # legend fontsize
plt.rc('figure', titlesize=15)  # fontsize of the figure title

x = np.linspace(0, 1, 1002)[1:-1]

plt.figure(1)
dist = beta(alpha_values[0],alpha_values[0])
plt.plot(x, dist.pdf(x), c=palette(1), label='Beta (%.1f,%.1f)' % (alpha_values[0], alpha_values[0]))
dist = beta(alpha_values[1],alpha_values[1])
plt.plot(x, dist.pdf(x), c=palette(2), label='Beta (%.1f,%.1f)' % (alpha_values[1], alpha_values[1]))
dist = beta(alpha_values[2],alpha_values[2])
plt.plot(x, dist.pdf(x), c=palette(3), label='Beta (%.1f,%.1f)' % (alpha_values[2], alpha_values[2]))
dist = beta(alpha_values[3],alpha_values[3])
plt.plot(x, dist.pdf(x), c=palette(6), label='Beta (%.1f,%.1f)' % (alpha_values[3], alpha_values[3]))

# Add legend and labels
plt.legend(loc='best')
plt.title("filler", color='black')
plt.xlabel("filler", color='black')
plt.ylabel("filler", color='black')
plt.xlim(0, 1)
plt.ylim(0, 3)

# Plot
plt.savefig('/Users/wildflowerlyi/Desktop/Plots/betas.png')

```
